```python
# 1. Organize dado pelo tipo
dadosTipo = dados.sort_values(by='Tipo')

dadosTipo
```

```python
# 2. Organize dado pelo Tipo de Falha

dadosTipoFalha = dados.sort_values(by='Tipo da Falha')

dadosTipoFalha
```

```python
# 3. Organize o dado pelo tipo e tipo de falha

dadosOrganizados = dados.sort_values(by=['Tipo', 'Tipo da Falha'])

dadosOrganizados
```

```python
# 4. Selecione apenas dados de máquina do tipo M

tipoM = dados[dados['Tipo'] == 'M']

tipoM
```

```python
# 5. Verifique quais sao as falhas para esse tipo de máquina
tipoM['Tipo da Falha'].value_counts()
```

```python
# 6. Selecione apenas dados de máquina do tipo M e H

tipoHM = dados[(dados['Tipo'] == 'M') | (dados['Tipo'] == 'H')]

tipoHM
```
